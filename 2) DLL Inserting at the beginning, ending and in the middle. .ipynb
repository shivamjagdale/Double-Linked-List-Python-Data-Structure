```python
# Actual Program
# Double Linked List Create and Display 

class Node:
    def __init__(self,data):
        self.data = data 
        self.prev = None
        self.next = None
        
class DLL:
    def __init__(self):
        self.head = None
        
    def insert_beginning(self, data):
        nb = Node(data)

        temp = self.head

        temp.prev = nb
        nb.next = temp
        self.head = nb
        
    def insert_ending(self, data):
        
        ne = Node(data)

        temp = self.head

        while temp.next is not None:
            temp = temp.next
        temp.next = ne
        ne.prev = temp
        
    def insert_middle(self, pos, data):
        nm = Node(data)

        temp = self.head

        for i in range(1,pos-1):
            temp = temp.next

        nm.prev = temp
        nm.next = temp.next
        temp.next.prev = nm
        temp.next = nm


        
    def display(self):
        if self.head is None:
            print("The list is empty")
        else:
            temp = self.head
            
            while temp:
                print(temp.data, "→", end=" ")
                temp = temp.next
        
l = DLL()

n1 = Node(10)
l.head = n1

n2 = Node(20)
n1.next = n2
n2.prev = n1

n3 = Node(30)
n2.next = n3
n3.prev = n2

l.insert_beginning(5)
l.insert_beginning(4)

l.insert_ending(40)
l.insert_ending(50)

l.insert_middle(3,25)
l.insert_middle(3,21)

l.display()
```
